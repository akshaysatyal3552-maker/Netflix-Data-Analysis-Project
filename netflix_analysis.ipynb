```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
%matplotlib inline
```

Load Dataset

```python

import pandas as pd

df = pd.read_csv('../data/netflix_titles.csv')
df.head()
```

Data Understanding

```python
print("Shape of Dataset:", df.shape)
df.info()
df.describe(include='all')
```

Ratings Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,6))
sns.countplot(
    y='rating',
    data=df,
    order=df['rating'].value_counts().index,
    palette="pastel"
)
plt.title('Distribution of Ratings')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.tight_layout()
```

Top Genres

```python
genres = df['listed_in'].str.split(',', expand=True).stack()
top_genres = genres.value_counts().head(10)

plt.figure(figsize=(8,4))
top_genres.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.tight_layout()

plt.savefig('../visuals/chart5_top_genres.png')
plt.show()
```

### Top 10 Genres
![Top Genres](visuals/chart5_top_genres.png)  
*Most popular genres on Netflix.

```python
# 🔹 Bookmark: Check listed_in column
df['listed_in'].head(10)
```

```python
import matplotlib.pyplot as plt

# Drop rows where 'listed_in' is NaN
genres_series = df['listed_in'].dropna().str.split(',', expand=True).stack()

# Strip extra spaces
genres_series = genres_series.str.strip()

# Count top 10 genres
top_genres = genres_series.value_counts().head(10)

# Plot
plt.figure(figsize=(8,5))
top_genres.plot(kind='bar', color='skyblue')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()

# Ensure visuals folder exists
import os
os.makedirs('../visuals', exist_ok=True)

# Save chart
plt.savefig('../visuals/chart5_top_genres.png')
plt.show()
```

### Content Growth Over Years
![Content Growth](visuals/chart3_content_growth.png)  
*Number of titles added to Netflix each year.

```python
df['country'].head(10)
```

```python
import matplotlib.pyplot as plt

# 🔹 Bookmark: Top Countries Distribution

# Drop NaN
countries_series = df['country'].dropna().str.split(',', expand=True).stack()
countries_series = countries_series.str.strip()  # Remove extra spaces

# Count top 10 countries
top_countries = countries_series.value_counts().head(10)

# Plot
plt.figure(figsize=(8,5))
top_countries.plot(kind='bar', color='orange')
plt.title('Top 10 Content-Producing Countries on Netflix')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()

# Ensure visuals folder exists
import os
os.makedirs('../visuals', exist_ok=True)

# Save chart
plt.savefig('../visuals/chart2_top_countries.png')
plt.show()
```
